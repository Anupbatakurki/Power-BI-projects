# Task 2 - Data Visualization




```python
#showing a correlation map for all numerics 
corr_matrix = df.corr(numeric_only = True)
plt.figure(figsize=(30, 15))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')  # Corrected 'coolwarn' to 'coolwarm'
plt.title('Correlation Map for Numeric Variables')
plt.show()
```

```python
#Overtime 
sns.countplot(data=df,x='OverTime')
plt.title("OverTime")
plt.show()
```

```python
#martail status
sns.countplot(data=df,x='MaritalStatus')
plt.title('Marital Status')
plt.show()
```

```python
#job role
plt.figure(figsize=(20,10))
sns.countplot(data = df ,y ="JobRole")
plt.title('Job Role')
plt.show()
```

```python
#Gender 
sns.countplot(data=df,x = 'Gender')
plt.title('Gender')
plt.show()
```

```python
#Educational Field
plt.figure(figsize=(20,10))
sns.countplot(data=df,x = 'EducationField')
plt.title("Educational Field")
plt.show()
```

```python
#Department
plt.figure(figsize=(20,10))
sns.countplot(data=df,x = 'Department')
plt.title("Department")
plt.show()
```

```python
#Department
plt.figure(figsize=(20,10))
sns.countplot(data=df,x = 'BusinessTravel')
plt.title("BusinessTravel")
plt.show()
```

```python
#Relationship between Overtime and Age 
plt.figure(figsize=(10,5))
sns.boxplot(data=df,x = 'OverTime', y = "Age")
plt.title("Relational between Overtime and age")
plt.show()
```

```python
#total working hours
sns.histplot(data= df,x='TotalWorkingYears' , bins= 10 ,kde = True)
plt.title('Total Working Years')
plt.show()
```

### Salary Analysis

```python
#monthly Income Disturbution
sns.histplot(data= df,x='MonthlyIncome' , bins=10 ,kde = True)
plt.title('Monthly Income')
plt.show()
```

```python
#Salary by Job Role
plt.figure(figsize=(10,5))
sns.boxplot(data=df,x='JobRole', y = 'MonthlyIncome')
plt.title('Salary by Job Role')
plt.xticks(rotation=90)
plt.show()
```
